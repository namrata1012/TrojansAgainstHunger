```python
#imports
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\koyan\\Downloads\\CBP2021.CB2100CBP-Data.csv")
```

```python
df
```

```python
codes = ['445110', '44529', '45231', '722513', '722515', '44512']
"""445110: "Supermarkets and Other Grocery (except Convenience) Stores." 

44529: "Other Specialty Food Stores,"

45231: "General Merchandise Stores," 
722513: "Limited-Service Restaurants,"
722515: "Snack and Nonalcoholic Beverage Bars." 

44512:  "Convenience Stores." 

"""
## It seems NAICS codes change too. Few NAICS codes which were mentioned in paper are not available in data, hence selecyed the relevant ones
df=df[df['NAICS2017'].isin(codes)]
```

```python
df['NAME']= df['NAME'].str[4:9] #To convert zipcodes to required format ie 'ZIP 90001 (Los Angeles, CA)=>90001'
```

```python
columns_to_drop = ['ESTAB_F','PAYANN', 'PAYANN_F', 'PAYANN_N', 'PAYANN_N_F', 'PAYQTR1', 'PAYQTR1_F', 'PAYQTR1_N', 'PAYQTR1_N_F', 'EMP', 'EMP_F', 'EMP_N', 'EMP_N_F']
df = df.drop(columns=columns_to_drop) #Dropping irrelavant columns
```

```python
df = df[df['EMPSZES'] == 1] #Selecting EMPSZES=1 since we need all establishments irrespective of theor employee size
df
```

```python
healthy_retailers =  ['445110', '44529', '45231']
unhealthy_retailers = ['722513', '722515', '44512']
# For calculating healthy establishments
df_healthy = df[df['NAICS2017'].isin(healthy_retailers)]

# Summing up number of establishments according to a pincode
df_healthy1 = df_healthy.groupby('NAME')['ESTAB'].sum().reset_index()
column_mapping = {
    'NAME': 'Zipcode',
    'ESTAB': 'HEALTHY'
}


df_healthy1 = df_healthy1.rename(columns=column_mapping)
print(df_healthy1) 

# For calculating unhealthy establishments
df_unhealthy = df[df['NAICS2017'].isin(unhealthy_retailers)]

# Summing up number of establishments according to a pincode
df_unhealthy1 = df_unhealthy.groupby('NAME')['ESTAB'].sum().reset_index()
column_mapping = {
    'NAME': 'Zipcode',
    'ESTAB': 'UNHEALTHY'
}

# Use the rename method to rename columns
df_unhealthy1 = df_unhealthy1.rename(columns=column_mapping)
print(df_unhealthy1)
```

```python
final_zip_df = pd.merge(df_healthy1, df_unhealthy1, on='Zipcode', how='outer')
final_zip_df
```

```python
datatoexcel = pd.ExcelWriter('ZipHealthyUnhealthy_2021.xlsx')
 
# write DataFrame to excel
final_zip_df.to_excel(datatoexcel)
 
# save the excel
datatoexcel.save()
```

```python
final_zip_df['Zipcode'] = final_zip_df['Zipcode'].astype(int) ##Converted since both columns on which dfs are being merged need to be of same datatype
result = pd.merge(final_zip_df, dt, on='Zipcode')
```
